## subplots  

`plt.subplots` 是 `matplotlib` 中一个非常常用的函数，用于创建一个包含子图（子绘图区域）的图形（Figure）对象和轴（Axes）对象。这个函数返回一个 `Figure` 对象和一个或多个 `Axes` 对象，方便用户在同一个图形中绘制多个子图。

fig, ax = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(10, 8))



### 主要参数介绍

1. **`nrows` 和 `ncols`**:
   - `nrows`：子图的行数。
   - `ncols`：子图的列数。
   - 示例：`plt.subplots(nrows=2, ncols=3)` 会创建一个 2 行 3 列的子图布局。

2. **`figsize`**:
   - 用于设置整个图形的大小，格式为 `(宽度, 高度)`（单位为英寸）。
   - 示例：`plt.subplots(figsize=(10, 8))` 会创建一个宽 10 英寸、高 8 英寸的图形。

3. **`sharex` 和 `sharey`**:
   - `sharex`：如果为 `True`，则子图共享相同的 X 轴刻度。
   - `sharey`：如果为 `True`，则子图共享相同的 Y 轴刻度。
   - 示例：`plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True)` 会让所有子图共享相同的 X 轴和 Y 轴。

4. **`squeeze`**:
   - 控制返回的 `Axes` 对象的形状。如果为 `True`（默认），那么当 `nrows` 或 `ncols` 为 1 时，返回的是一个 `Axes` 对象而不是一个包含单个 `Axes` 对象的数组。
   - 示例：`plt.subplots(nrows=1, ncols=1, squeeze=False)` 会返回一个 1x1 的数组而不是单个 `Axes` 对象。

5. **`subplot_kw`**:
   - 一个字典，用于传递子图 (`subplot`) 的关键字参数。可以控制每个 `Axes` 的属性。
   - 示例：`plt.subplots(nrows=2, ncols=2, subplot_kw={'projection': 'polar'})` 会创建极坐标的子图。

6. **`gridspec_kw`**:
   - 用于传递与网格布局（`GridSpec`）相关的关键字参数。比如可以控制子图之间的宽度和高度比等。
   - 示例：`plt.subplots(nrows=2, ncols=2, gridspec_kw={'width_ratios': [2, 1]})` 会创建两个子图，其中第一列的子图宽度是第二列子图的两倍。

7. **`constrained_layout`**:
   - 如果为 `True`，则会自动调整子图布局，以避免子图之间的重叠。
   - 示例：`plt.subplots(nrows=2, ncols=2, constrained_layout=True)` 会自动调整子图间距，防止重叠。

8. **`layout`**:
   - 用于设置子图的布局方式，值为 `"constrained"` 或 `"tight"`。`"constrained"` 是 `constrained_layout` 参数的简写。

### 返回值
`plt.subplots` 返回两个对象：
- **`fig`**: `Figure` 对象，代表整个图形容器。
- **`ax`**: `Axes` 对象或 `Axes` 对象的数组，表示子图区域。

````python
import matplotlib.pyplot as plt

# 创建一个2x2的子图布局，所有子图共享X轴和Y轴
fig, ax = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(10, 8))

# 读取图像数据
img1 = mpimg.imread('test_img.jpg')  # 图像 1
img2 = img1  # 图像 2
img3 = img1  # 图像 3
img4 = img1  # 图像 4

# 在每个子图中显示图像
ax[0, 0].imshow(img1)
ax[0, 0].set_title('Image 1')
ax[0, 0].axis('off')  # 关闭坐标轴

ax[0, 1].imshow(img2)
ax[0, 1].set_title('Image 2')
ax[0, 1].axis('off')  # 关闭坐标轴

ax[1, 0].imshow(img3)
ax[1, 0].set_title('Image 3')
ax[1, 0].axis('off')  # 关闭坐标轴

ax[1, 1].imshow(img4)
ax[1, 1].set_title('Image 4')
ax[1, 1].axis('off')  # 关闭坐标轴

# 调整子图间距以避免重叠
plt.tight_layout()

# 显示图形
plt.show()
# 显示图形
plt.show()
````

**`plt.subplots`当存在多个子图时，返回`ax`数据类型为数组；只有一个子图时`ax`为单一的 `Axes`对象**

````python
fig, ax = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(10, 8))
fig1, ax1 = plt.subplots(nrows=1, ncols=1, sharex=True, sharey=True, figsize=(10, 8))
fig2, ax2 = plt.subplots(nrows=1, ncols=1, sharex=True, sharey=True, figsize=(10, 8),squeeze=False)

ax[0][0].set_xlabel("1")

type(ax),ax.shape,type(ax1),type(ax2),ax2.shape,len(ax2.shape),ax2.ndim
````

`plt.subplot` 是 `matplotlib` 中的一个函数，用于在一个绘图区域内创建多个子图（subplots）。它允许你在同一个 `Figure` 中排列多个图形，以便进行对比和展示。这个函数的三个主要参数决定了子图的布局和具体绘制的位置。

### `plt.subplot` 函数的参数

`plt.subplot(nrows, ncols, index)` 中的三个参数分别为：

1. **`nrows`（行数）**：
   - 指定子图的行数，即整个绘图区域被划分为多少行。

2. **`ncols`（列数）**：
   - 指定子图的列数，即整个绘图区域被划分为多少列。

3. **`index`（索引）**：
   - 指定当前子图的位置，索引从 1 开始，从左到右，从上到下按顺序排列。
   - 例如，如果你设置了 `nrows=2` 和 `ncols=3`，则子图的索引范围为 1 到 6，排列如下：
     ```
     | 1 | 2 | 3 |
     | 4 | 5 | 6 |
     ```

### 具体示例

假设你想在同一个图形中绘制三个不同的图，可以按以下方式使用 `plt.subplot`：


### 解释

- **`plt.subplot(2, 2, 1)`**：创建一个 2x2 的网格布局，选择第 1 个位置的子图。
- **`plt.subplot(2, 2, 2)`**：选择第 2 个位置的子图。
- **`plt.subplot(2, 2, 3)`**：选择第 3 个位置的子图。
- 这样，你就可以在同一个 `Figure` 中排列和显示多个图形。

### 可选参数
`plt.subplot` 还支持可选参数，用于控制更多细节：
- **`projection`**：指定子图的投影类型，例如 `'polar'` 用于极坐标图。
- **`polar`**：等价于 `projection='polar'`，直接生成极坐标图。
- **`xscale`, `yscale`**：设置子图的 X 轴或 Y 轴的缩放类型，如 `'log'`、`'linear'` 等。
- **`aspect`**：控制子图的长宽比。

### 总结
`plt.subplot` 是一个非常灵活的函数，适用于简单的子图布局。通过指定行数、列数和索引，你可以在同一个 `Figure` 中创建多个子图，方便对比和展示不同的数据。

````python
import matplotlib.pyplot as plt

# 创建一个 2x2 的子图布局，绘制第 1 个子图
plt.subplot(2, 2, 1)
plt.imshow(img1)
plt.title('Plot 1')

# 绘制第 2 个子图
plt.subplot(2, 2, 2)
plt.imshow(img1)
plt.title('Plot 2')

# 绘制第 3 个子图
plt.subplot(2, 2, 3)
plt.imshow(img1)
plt.title('Plot 3')

# 显示所有子图
plt.tight_layout()  # 自动调整子图间距，防止标题和标签重叠
plt.show()
````

### 格式字符串解析    
`'-'`:  
说明: 这是一个简单的实线。    
颜色: 默认颜色（通常是**黑色**，除非指定了颜色）。    
示例: 用于绘制一条实线。

线型: `'--'` 表示虚线。    
示例: 用于绘制一条粉红色的虚线。  

`'g-.'`:  
说明: 这是一个**绿色的点划线**。  
颜色: `'g'` 表示绿色（green）。  
线型: `'-.'` 表示点划线。  
示例: 用于绘制一条绿色的点划线。  

`'r:'`:  
说明: 这是一个**红色的点线**。  
颜色: `'r'` 表示红色（red）。  
线型: `':'` 表示点线。  
示例: 用于绘制一条红色的点线。  

`plot` 是 `matplotlib.pyplot` 模块中的一个函数，用于绘制二维图形。它是最常用的绘图函数之一，用于绘制线图、散点图等。下面是对 `plot` 函数的详细解释和示例。

### 基本语法

```python
matplotlib.pyplot.plot(*args, **kwargs)
```

- **`*args`**: 位置参数，用于传入绘图数据。通常是 x 和 y 数据。
- **`**kwargs`**: 关键字参数，用于设置图形的属性（如颜色、线型、标签等）。

### 常用参数

- **`x`**: x 轴数据。可以是列表、数组或其他可迭代对象。
- **`y`**: y 轴数据。可以是列表、数组或其他可迭代对象。
- **`color`**: 绘图颜色。可以是颜色名称、RGB 代码或颜色缩写。
- **`linestyle`**: 线型。常见的有 `'-'`（实线）、`'--'`（虚线）、`'-.'`（点划线）、`':'`（点线）。
- **`marker`**: 标记样式。常见的有 `'.'`（点）、`'o'`（圆圈）、`'s'`（方块）、`'^'`（三角形）。
- **`label`**: 图例标签，用于图例中显示数据的描述。

### 示例

#### 基本线图

```python
import matplotlib.pyplot as plt

# 数据
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# 绘制线图
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

#### 自定义颜色和线型

```python
import matplotlib.pyplot as plt

# 数据
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# 绘制线图，设置颜色和线型
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.title("Custom Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

#### 多条线图

```python
import matplotlib.pyplot as plt

# 数据
x = [0, 1, 2, 3, 4]
y1 = [0, 1, 4, 9, 16]
y2 = [0, 1, 8, 27, 64]

# 绘制多条线
plt.plot(x, y1, label='y = x^2', color='blue', linestyle='-')
plt.plot(x, y2, label='y = x^3', color='red', linestyle='--')
plt.title("Multiple Lines Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.show()
```

#### 散点图（通过设置 `marker`）

```python
import matplotlib.pyplot as plt

# 数据
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# 绘制散点图
plt.plot(x, y, 'ro')  # 'ro' 表示红色圆点
plt.title("Scatter Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

#### 设置线宽和透明度

```python
import matplotlib.pyplot as plt

# 数据
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# 绘制线图，设置线宽和透明度

plt.title("Line Plot with Line Width and Alpha")
plt.xlabel("X axis")
plt.ylabel("Y axis")

```

### 总结

`plot` 函数是 `matplotlib` 中非常重要的绘图函数，用于绘制线图、散点图等。它支持丰富的自定义选项，包括颜色、线型、标记样式、线宽等，适用于各种数据可视化需求。通过灵活使用 `plot` 函数，你可以创建清晰、精美的图形来展示数据。

````python
import matplotlib.pyplot as plt

# 数据
x = [0, 1, 2, 3, 4]
y = [(-1,0), (0,1), (3,4), (8,9), (15,16)]

# 绘制线图
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
````
